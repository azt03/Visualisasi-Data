# VISUALISASI DATA

Hubungan antara status mahasiswa dan IPK Semester 1-6

Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load Dataset

```python
data = pd.read_csv('Data Mahasiswa.csv')
df = pd.DataFrame(data)
data
```

Membuat Table Pivot

```python
pivot_table = df.pivot_table(values='IPK SEMESTER 1-6', 
                             index='STATUS MAHASISWA', 
                             aggfunc=['mean', 'min', 'max'])
print(pivot_table)
```

Visualisasi Heatmap

```python
# Membuat heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table['mean'], annot=True, cmap='YlGnBu')
plt.title('Rata-rata IPK Berdasarkan Status Mahasiswa')
plt.ylabel('Status Mahasiswa')
plt.xlabel('Rata-rata IPK')
plt.show()
```

Visualisasi Countplot

```python
# Membuat countplot untuk Status Mahasiswa dan Jenis Kelamin
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='STATUS MAHASISWA', hue='JENIS KELAMIN')
plt.title('Status Mahasiswa Berdasarkan Jenis Kelamin')
plt.xlabel('Status Mahasiswa')
plt.ylabel('Jumlah')
plt.show()
```
